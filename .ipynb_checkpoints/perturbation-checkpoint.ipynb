```python
import numpy as np
import pandas as pd
```

```python
# import training data
train_comments = pd.read_csv('train_comments.csv')
```

How to create one perturbation:

```python
# dataframe column to numpy array
tox_np = train_comments.binary_tox.values
```

```python
print(tox_np.tolist())
# tox_np.tolist()
```

```python
list_perturbation = []
# generate a random number (between 0 and 1) for each comment


# if random number is greater than 0.5, replace value in array with a random integer from [0, 1]
for j in range(10):
    rand = np.random.random(len(tox_np))
    tox_tmp = tox_np
    for i in range(len(tox_np)):
        if rand[i] >= 0.5:
            tox_tmp[i] = np.random.randint(2)
    list_perturbation.append(tox_tmp)
```

```python
# add perturbed numpy array to dataframe
train_comments['new_binary_tox'] = tox_np
```
